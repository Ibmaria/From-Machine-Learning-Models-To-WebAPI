```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import joblib
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
```

```python
df =pd.read_csv('spam.csv',encoding="latin-1")
df.head()
```

```python
# dropping unnecesary columns.
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)

# changing column name
df.columns = ['target', 'text']

# changing label to 0 and 1
df['target'] = df['target'].map({'ham': 0, 'spam': 1})
```

```python
df['text'][2]
```

```python
# Feature extraction with CountVectorizer
X = df['text']
y = df['target']
cv = CountVectorizer()
cv_fit = cv.fit(X)
X = cv.transform(X) # Fit the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
pickle.dump(cv_fit, open('model_pickel.pkl', 'wb'))
```

```python
# saving the vector
joblib.dump(cv_fit, 'model_pickel.pkl')
```

```python

#Naive Bayes Classifier
clf = MultinomialNB()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
pickle.dump(clf, open('NaiveBayesClassifier.sav', 'wb'))
```

```python

joblib.dump(clf, 'NaiveBayesClassifier.sav')
```

```python
#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=23)      
clr_rf = clf_rf.fit(X_train,y_train)
```

```python
clf_rf.score(X_test,y_test)
y_pred = clf_rf.predict(X_test)
print(classification_report(y_test, y_pred))
```
